# SIGNAL DATA EXTRACTION
--------------------

Extraction for 506889D3-2F1F-4C3E-8C8D-3F8A1F4E1E1C (signal)

```python
import math

# Define the vectors to hold the jet data (ensure these types match those used in the TTree)
jetPt = ROOT.std.vector('float')()
jetEta = ROOT.std.vector('float')()
jetPhi = ROOT.std.vector('float')()

# Set the branch addresses
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fPt", jetPt)
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fEta", jetEta)
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fPhi", jetPhi)

# Prepare a new file to store the processed data
outputFile = ROOT.TFile("processed_data_signal.root", "RECREATE")
outputTree = ROOT.TTree("JetData", "Processed data for ML")

# Define variables to hold the processed data
px = ROOT.std.vector('float')()
py = ROOT.std.vector('float')()
pz = ROOT.std.vector('float')()

# Create branches in the output tree for the processed data
outputTree.Branch("px", px)
outputTree.Branch("py", py)
outputTree.Branch("pz", pz)

# Loop over the events in the original tree
for i in range(tree.GetEntries()):
    if i % 100 == 0: 
        print("Running... Processed {} events".format(i))
   
    tree.GetEntry(i)
    
    # Clear the vectors for the new event
    px.clear()
    py.clear()
    pz.clear()
    
    # Process the first 15 jets
    num_jets = min(15, jetPt.size())
    for j in range(num_jets):
        px.push_back(jetPt.at(j) * math.cos(jetPhi.at(j)))
        py.push_back(jetPt.at(j) * math.sin(jetPhi.at(j)))
        pz.push_back(jetPt.at(j) * math.sinh(jetEta.at(j)))

    # Fill the output tree with the processed event data
    outputTree.Fill()

# Write the output tree to file
outputFile.Write()
outputFile.Close()
```

Extraction for 2888029DF-2F1F-4C3E-8C8D-3F8A1F4E1E1C (signal)

```python
import math

# Open the input ROOT file
file = ROOT.TFile.Open("288029DF-A656-E611-B735-6C3BE5B5C0B0.root")

# Get the Events tree from the input file
tree = file.Get("Events")

# Define the vectors to hold the jet data
jetPt = ROOT.std.vector('float')()
jetEta = ROOT.std.vector('float')()
jetPhi = ROOT.std.vector('float')()

# Set the branch addresses
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fPt", jetPt)
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fEta", jetEta)
tree.SetBranchAddress("patJets_slimmedJets__PAT.obj.m_state.p4Polar_.fCoordinates.fPhi", jetPhi)

# Open the output ROOT file in "UPDATE" mode
outputFile = ROOT.TFile("processed_data_signal.root", "UPDATE")

# Check if the tree already exists in the file
outputTree = outputFile.Get("JetData")
if not outputTree:
    # Create a new TTree if it does not exist
    print("THIS SHOULDN'T HAPPEN")
    outputTree = ROOT.TTree("JetData", "Processed data for ML")

# Define variables to hold the processed data
px = ROOT.std.vector('float')()
py = ROOT.std.vector('float')()
pz = ROOT.std.vector('float')()

# Create branches in the output tree if it's new
if outputFile.GetListOfKeys().Contains("JetData"):
    outputTree.SetBranchAddress("px", px)
    outputTree.SetBranchAddress("py", py)
    outputTree.SetBranchAddress("pz", pz)
else:
    outputTree.Branch("px", px)
    outputTree.Branch("py", py)
    outputTree.Branch("pz", pz)

# Loop over the events
entries = tree.GetEntries()
print("Processing {} events...".format(entries))
for i in range(entries):
    if i % 100 == 0:  # Print a message every 100 iterations
        print("Running... Processed {} events".format(i))

    tree.GetEntry(i)
    px.clear()
    py.clear()
    pz.clear()
    
    num_jets = min(15, jetPt.size())
    for j in range(num_jets):
        px.push_back(jetPt.at(j) * math.cos(jetPhi.at(j)))
        py.push_back(jetPt.at(j) * math.sin(jetPhi.at(j)))
        pz.push_back(jetPt.at(j) * math.sinh(jetEta.at(j)))

    outputTree.Fill()

# Write the updated data and close the file
outputFile.Write("", ROOT.TObject.kOverwrite)
outputFile.Close()

print("Data processing complete.")
```
